```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import matplotlib
from matplotlib import rc
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

rc('text', usetex=False)
sns.set_context('paper', font_scale=2.0)
sns.set_style('whitegrid')
```

```python
raw = pd.read_csv("tlb.debug", header=None)
raw.columns = ['name', 'tlb', 'page', 'time']
raw.time = (raw.time - raw.time[0])*1e6
raw
```

```python
group_by_page = raw.groupby('page').agg({'tlb':'nunique'}).reset_index()
group_by_page = group_by_page.sort_values('tlb', ascending=False).reset_index()
group_by_page
```

```python
pages = group_by_page['page']
pages
```

```python
selected_pages = [pages[0], pages[3], pages[4], pages[6], pages[7], pages[8], pages[9]]
tlbs = [63, 47, 35, 25, 12]
data = raw[raw['page'].isin(selected_pages)]
data = data[(data.time > 49150) & (data.time < 49500)]
data = data[data['tlb'].isin(tlbs)]
data = data.sort_values(by='time')
data
```

```python
data.tlb.unique()
```

```python
fig, ax = plt.subplots(figsize=(10, 2))

sns.lineplot(data=data, x='time', y='tlb', hue='page', marker="o", 
             palette=sns.color_palette(n_colors=len(selected_pages)), 
             ax=ax, 
             markersize=8, markeredgewidth=1.5, linewidth=1.5)
    
handles, labels = ax.get_legend_handles_labels()
labels = ["Pages", "A", "B", "C", "D", "E", "F", "G"]
ax.legend(handles, labels, 
          frameon=False,
          fancybox=None, 
          columnspacing=1.5,
          handlelength=1,
          facecolor=None, edgecolor=None, 
          bbox_to_anchor=(0.47, 1.33), 
          loc=9, ncol=8)

ax.set_xlim([49150, 49500])
ax.set_yticks(tlbs)
ax.set_xlabel("Kilo-Cycles")
ax.set_ylabel("TLB Index")


plt.savefig('tlb_pattern.pdf', bbox_inches='tight')
plt.savefig('tlb_pattern.png', dpi=600, bbox_inches='tight')
plt.show()
plt.close()
```
